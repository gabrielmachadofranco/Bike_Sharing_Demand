# Análise da variação anual

```python
train ['year_month'] = pd.to_datetime(train['datetime']).dt.strftime('%Y-%m')
by_year_month = train[['count','year_month']].groupby('year_month').mean().reset_index()
plt.figure(figsize=(20,12))
plt.plot(by_year_month['year_month'],by_year_month['count'])
plt.xticks(rotation=60)
plt.title("Variação dos aluguéis por mês", size = 20)
```

# Modelo preditivo para cálculo da demanda

```python
# Para realizar a previsão da demanda por bicicletas foram testados dois modelos de machine learning: Regressão Linear e
# Random Forest
```

```python
X = train.drop(['count', 'registered', 'casual','datetime','year_week','year_month'], axis = 'columns')
X.head()
```

```python
y = train['count']
y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(X,y,test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics


# Train machine learn
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)

rf_reg = RandomForestRegressor()
rf_reg.fit(x_train, y_train)
```

```python
# Test machine learn
test_pred_lin = lin_reg.predict(x_test)
test_pred_rf = rf_reg.predict(x_test)

# Machine Learn result analyze
r2_lin = metrics.r2_score(y_test, test_pred_lin)
mse_lin = metrics.mean_squared_error(y_test, test_pred_lin)

print (f'R² - regressão linear: {r2_lin}')
print (f'R² - Random Forest: {r2_rf}')


r2_rf = metrics.r2_score(y_test, test_pred_rf)
mse_rf = metrics.mean_squared_error(y_test, test_pred_rf)


print (f'MSE - regressão linear: {mse_lin}')
print (f'MSE - Random Forest: {mse_rf}')
```

```python
# A partir do resultado da análise de desempenho dos dois modelos, verificou-se que o Random Forest (RF) obteve um melhor 
# resultado pois teve o maior R² e o menor MSE. Portanto, o RF foi escolhido para realizar a previsão da demanda.
```
